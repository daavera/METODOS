```python
import numpy as np
import matplotlib.pyplot as plt
```

# DERIVADAS

```python
A=-5
B=5
N=1001
x = np.linspace(A,B,N)
h = x[1]-x[0]
```

```python
def f(x):
    y = np.exp(-x**2)
    return y
y = f(x)
dy = -2*x*np.exp(-x**2)
dy2= -2*np.exp(-x**2) + 4*(x**2)*np.exp(-x**2)
```

## Forward difference

```python
def derivada_f(y,h):
    derivada_f = (f(x+h)-f(x))/h
    return derivada_f
f1, ax = plt.subplots(2, sharex=True)
ax[0].plot(x,y, c = 'red', label="funcion")
ax[0].plot(x,derivada_f(y,h), label ="derivada")
ax[0].legend(loc='upper right')
ax[0].set_title('Funciones')
error =abs(derivada_f(y,h)-(dy))/h
ax[1].plot(x,error, c = 'green')
ax[1].set_title('Errores')
plt.show()
```

## Central difference

```python
def derivada_c(y,h):
    derivada_c = (f(x+h/2)-f(x-h/2))/h
    return derivada_c


f1, ax = plt.subplots(2, sharex=True)
ax[0].plot(x,y, c = 'red', label ='funcion')
ax[0].plot(x,derivada_c(y,h), label = 'derivada')
ax[0].legend(loc='upper right')
ax[0].set_title('Funciones')
error =abs(derivada_c(y,h)-(dy))/(h**2)
ax[1].plot(x,error, c = 'green')
ax[1].set_title('Errores')
plt.show()
```

## Second derivate

```python
def derivada2(y,h):
    derivada2 = (f(x+h)+f(x-h)-(2*y))/(h**2)
    return derivada2

f1, ax = plt.subplots(2, sharex=True)
ax[0].plot(x,y, c = 'red', label ='funcion')
ax[0].plot(x,derivada2(y,h), label = 'derivada')
ax[0].legend(loc='upper right')
ax[0].set_title('Funciones')
error =abs(derivada2(y,h)-(dy2))/h**2
ax[1].plot(x,error, c = 'green')
ax[1].set_title('Errores')
plt.show()
```

## Extrapolated

```python
def derivada_e(y,h):
    derivada_e = (4*derivada_c(y,h/2)-derivada_c(y,h))/3
    return derivada_e
f1, ax = plt.subplots(2, sharex=True)
ax[0].plot(x,y, c = 'red', label ='funcion')
ax[0].plot(x,derivada_e(y,h), label = 'derivada')
ax[0].legend(loc='upper right')
ax[0].set_title('Funciones')
error =abs(derivada_e(y,h)-(dy))/h**2
ax[1].plot(x,error, c = 'green')
ax[1].set_title('Errores')
plt.show()
```

# COMAPARACIÓN ERRORES

```python
def prom_integral(y,B,A,N):
    ancho = 1/(N-1)
    odd = np.sum(y[1:-1:2])*4/3
    pair = np.sum(y[2:-1:2])*2/3
    h = (y[0]+y[-1])/3 + odd + pair
    return(h*ancho)
```

```python
h_s = np.logspace(-9,3,200)
err_f = []
err_c = []
err_e = []
for H in h_s:
    #Forward
    f_f = abs(derivada_f(y,H) - dy)
    err_f.append(prom_integral(f_f,B,A,N))
    #Center
    f_c = abs(derivada_c(y,H) - dy)
    err_c.append(prom_integral(f_c,B,A,N))
    #Extrapolated
    f_e = abs(derivada_e(y,H) - dy)
    err_e.append(prom_integral(f_e,B,A,N))  
```

```python
fig = plt.figure(figsize =(9,6))
plt.plot(h_s,err_f, label='Forward')
plt.plot(h_s,err_c, label = 'Central')
plt.plot(h_s,err_e, label='Extrapolated')
plt.xscale('log')
plt.yscale('log')
plt.legend(loc='upper left')
plt.show()
```
