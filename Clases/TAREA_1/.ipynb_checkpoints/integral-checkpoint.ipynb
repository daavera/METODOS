```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
def f (X):
    y = 0
    for i in range(10):
        x = X[i]
        y += x
    return y**3
def random(n_random, X):
    RANDOM =[]
    for z in range(11):
        r = np.random.rand(n_random)
        r = r.tolist()
        RANDOM.append(r)
    RANDOM = np.array(RANDOM)
    for j in range(10):
        RANDOM[j] = RANDOM[j] * np.max(X[j])-np.min(X[j]) + np.min(X[j])
    random_y = RANDOM[10] * np.max(y)-np.min(y) + np.min(y)
    return random_y, RANDOM
    
```

```python
a = 0
b = 2
N = 100
X = []
for i in range(10):
    x = np.linspace(a,b,N)
    X.append(x)
y = f(X)
```

```python
n_random = 10000
RANDOM =[]
for i in range(11):
    r = np.random.rand(n_random)
    r = r.tolist()
    RANDOM.append(r)
RANDOM = np.array(RANDOM)
for i in range(10):
    RANDOM[i] = RANDOM[i] * np.max(X[i])-np.min(X[i]) + np.min(X[i])
random_y = RANDOM[10] * np.max(y)-np.min(y) + np.min(y)
```

```python
delta = f(RANDOM) - random_y
inside = np.where(delta>0.0)
```

# CALCULO DE LA INTEGRAL 

```python
#Definicin de los limites
a = 0
b = 2
N = 1000

#Definicion de la funcin 
X = []
for i in range(10):
    x = np.linspace(a,b,N)
    X.append(x)
y = f(X)

#Integracin por Monte Carlo
n_random = 10000
y_rand , x_rand = random(n_random, X)

delta = f(x_rand) - y_rand
inside = np.where(delta>0.0)
#Intervalo
interval_I = ((b - a) ** 10) * (np.max(y)-np.min(y))
I  = interval_integral * (np.size(inside)/(1.0*np.size(y_rand)))
I
```

# INTEGRAL 20

```python
#Definicin de los limites
a = 0
b = 2
N = 1000

#Definicion de la funcin 
X = []
for i in range(10):
    x = np.linspace(a,b,N)
    X.append(x)
y = f(X)

#Lista para las repeticiones
integral = []
n_random = 10000
#Integrar 20 veces
for i in range(20):
    #Integracin con Monte Carlo
    y_rand , x_rand = random(n_random, X)
    delta = f(x_rand) - y_rand
    inside = np.where(delta>0.0)
    #Intervalo
    interval_I = ((b - a) ** 10) * (np.max(y)-np.min(y))
    I  = interval_integral * (np.size(inside)/(1.0*np.size(y_rand)))
    integral.append(I)
integral = np.array(integral)
prom_integral = np.sum(integral)/len(integral)
prom_integral
```

# VARIACION DE PUNTOS

```python
#Definicin de los limites
a = 0
b = 2
N = 100
n_random = np.arange(2,8193,20)

n_integral = []
#Definicion de la funcin 
X = []
for i in range(10):
    x = np.linspace(a,b,N)
    X.append(x)
y = f(X)

for n in n_random:
    integral = []
    for i in range(20):
    #Integracin con Monte Carlo
        y_rand , x_rand = random(n, X)
        delta = f(x_rand) - y_rand
        inside = np.where(delta>0.0)
    #Intervalo
        interval_I = ((b - a) ** 10) * (np.max(y)-np.min(y))
        I  = interval_integral * (np.size(inside)/(1.0*np.size(y_rand)))
        integral.append(I)
    integral = np.array(integral)
    prom_integral = np.sum(integral)/len(integral)
    n_integral.append(prom_integral)
n_integral = np.array(n_integral)
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.ticklabel_format(style='sci',scilimits=(0,6),axis='both')
ax.plot(n_random,np.full((1,len(n_random)),1126400)[0], linestyle='--', linewidth = 0.67, label='Integral Analtica', c='g')
ax.plot(n_random,n_integral, linewidth = 1, label='Integral numrica')
plt.grid(True)
plt.ylabel('Valor integral')
plt.xlabel('N')
plt.title('Integral promedio VS. Cantidad de nmeros aleatorios')
ax.fill_between(n_random, np.full((1,len(n_random)),1126400)[0], n_integral, facecolor='red', alpha=0.1)
ax.legend(loc='best', fancybox=True, framealpha=0.5)
#plt.savefig('num_integral.pdf')
plt.show()
```

```python
int_analitica = 1126400
error = (abs(n_integral-int_analitica)/int_analitica) * 100
sq_N = n_random**(-1/2)

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)
ax.plot(sq_N,error, linewidth = 0.67)
plt.grid(True)
plt.ylabel('Error relativo porcentual')
plt.xlabel('# de nmeros aleatorios')
plt.title(r'Error VS. $\frac{1}{\sqrt{N}}$',fontsize=20)
ax.set_yscale('log')
ax.set_xscale('log')
plt.tight_layout()
#plt.savefig('err_integral.pdf')
plt.show()
```
