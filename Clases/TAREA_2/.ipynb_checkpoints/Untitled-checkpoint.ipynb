```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
```

```python
ss_DO, DO = wav.read('Do.wav')
t_DO = np.linspace(0,np.size(DO)/ss_DO,np.size(DO))
ss_SOL ,SOL = wav.read('Sol.wav')
t_SOL = np.linspace(0,np.size(SOL)/ss_SOL,np.size(SOL))
```

```python
DO_F = np.fft.fft(DO)
f_DO = np.fft.fftfreq(np.size(DO),1/ss_DO)
SOL_F = np.fft.fft(SOL)
f_SOL = np.fft.fftfreq(np.size(SOL),1/ss_SOL)
```

```python
#Transformada de Fourier propia
def T_fourier (y):
    t = np.arange(np.size(y))
    N = np.size(y)
    n = np.arange(0,N,1).reshape(-1,1)
    R = np.exp(-2*np.pi*1j*n*t/N)
    return np.dot(R,y)
```

```python
#FILTROS
###filtro 1
def f_mayor(f_I,A_I):
    A = A_I.copy()
    F = f_I.copy()
    A_max = np.max(abs(A))
    i_max = np.where(abs(A) > A_max-1)[0]
    for i in i_max:
        index = np.arange(i-60,i+60,1)
        F[index] = 0
        A[index] = 0
    return F,A
###filtro 2
def f_1000HZ(f_I,A_I):
    A = A_I.copy()
    F = f_I.copy()
    index = np.where(abs(F) > 1000)[0]
    A[index] = 0
    return F,A
```

```python
F1, A1 = f_mayor(f_DO,DO_F)
F2, A2 = f_1000HZ(f_DO,DO_F)
f1 , ax = plt.subplots(3,1,figsize=(10,6), sharex=True, sharey=True)
ax[0].plot(f_DO, abs(DO_F))
ax[1].plot(F1,abs(A1), c='g')
ax[2].plot(F2,abs(A2), c='y')
ax[2].ticklabel_format(style='sci',scilimits=(1,6),axis='both')

ax[0].set_title('Transformada de Fourier')
ax[1].set_title('Filtro para la frecuencia con mayor amplitud')
ax[2].set_title('Filtro para frecuencias superiores a 1000Hz')

ax[1].set_ylabel('Amplitudes')
ax[2].set_xlabel('Frecuencia (Hz)', fontsize=12)
plt.show()
```

```python
#DO to SOL
index = np.where(f_DO == 260)
YA = False
new_ssDO = ss_DO
new_fDO = 0
while not(YA):
    new_fDO = np.fft.fftfreq(np.size(DO),1/new_ssDO)
    if(new_fDO[index] > 390 and new_fDO[index] < 392 ):
        YA = not(YA)
    else:
        new_ssDO = new_ssDO+1
f2 = plt.figure(figsize=(10,6))
plt.plot(new_fDO,abs(DO_F))
plt.plot(f_SOL,abs(SOL_F))
plt.show()
```

```python
DO_pico = np.fft.ifft(A1)
DO_pasabajos = np.fft.ifft(A2)
DO_pico
```

```python
wav.write('DO_pico.wav', ss_DO, abs(DO_pico))
wav.write('DO_pasabajos.wav', ss_DO, abs(DO_pasabajos))
wav.write('DoSol.wav', new_ssDO, DO)
```
