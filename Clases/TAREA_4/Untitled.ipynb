```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("map_data.txt")
```

```python
coord_x = 150
coord_y = 300

def circulo(xC,yC,r):
    t = np.linspace(0,2*np.pi,1000)
    x = r*np.cos(t) + xC
    y = r*np.sin(t) + yC
    return x,y
x,y = circulo(coord_x,coord_y,100)

plt.imshow(data)
plt.scatter(coord_x,coord_y)
plt.plot(x,y)
plt.show()
```
