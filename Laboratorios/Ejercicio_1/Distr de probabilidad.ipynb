```python
import matplotlib.pyplot as plt
from scipy.stats import expon, norm
import numpy as np
plt.style.use('ggplot')
```

# Distribución Exponencial

```python
n = np.random.exponential(10,1000)
fig = plt.figure(figsize=(7,4))
a = np.histogram(n, bins=20)
plt.hist(n,normed=True, bins=20, alpha = 0.8)
lamb=1/expon.fit(n)[1]
print(lamb)
x=np.linspace(np.min(a[1]),np.max(a[1]),1000)
plt.plot(x,lamb*np.exp(-lamb*x))
plt.show()
```

# Distribución Gaussiana

```python
a =[]
#Datos
for i in range(1000):
    n = np.random.exponential(10,1000)
    a.append(np.sum(n))
a = np.array(a)
hist = np.histogram(a, bins=20)
#Ajuste
x=np.linspace(np.min(hist[1]),np.max(hist[1]),1000)
mu, sig = norm.fit(a)
p = norm.pdf(x, mu, sig)
#Grafica
fig = plt.figure(figsize=(7,4))
plt.hist(a, normed=True, bins=20, alpha=0.8)
plt.plot(x,p, linewidth=2)
plt.show()
```
